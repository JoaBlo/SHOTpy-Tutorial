# SHOTpy Tutorial: Using SHOT from Python

**The Supporting Hyperplane Optimization Toolkit (SHOT)** is an open-source solver for mixed-integer nonlinear programming (MINLP) problems that uses a combination of polyhedral outer approximation, extended supporting hyperplanes, and other techniques. This tutorial provides a comprehensive guide to using **SHOTpy**, the Python interface for the SHOT solver.

SHOTpy exposes SHOT’s internal callbacks to Python, enabling users to customize solver behavior, experiment with new heuristics and/or polyhedral approximation strategies. Beyond solving, SHOT can thus serve as a **flexible platform for optimization research and experimentation**.

The Python interface also provides access to SHOT’s internal functionality, including loading models, configuring solver options, inspect detailed solution information, retrieve the complete solution pool with ease, and more. **Thus, the interface enables new ways of utilizing SHOT in advanced workflows and applications.**




The tutorial is divided into **three different notebooks**:
1. [SHOTpy_Tutorial_01_getting_started.ipynb](#SHOTpy_Tutorial_01_getting_started.ipynb)
    - Installation
    - Creating variables
    - Solving LP/MILP/MI(QC)QP/MINLP problems (linear, quadratic, exponential, logarithmic, trigonometric, square root, power function, monomial, signomial terms and mixed expressions)
    - Print problem structure and retrieving solution details
2.  [SHOTpy_Tutorial_02_solver_features.ipynb](#SHOTpy_Tutorial_02_solver_features.ipynb)
    - Reading problems from file
    - Configuring solver settings
    - Solver callbacks (`NewPrimalSolution`, `PrimalSolutionCandidateSelection`, `UserTerminationCheck`, `ExternalDualBound`, `ExternalPrimalSolution`, `ExternalHyperplaneSelection`) 

3. [SHOTpy_Tutorial_03_advanced_usage_and_extensions.ipynb](#SHOTpy_Tutorial_03_advanced_usage_and_extensions.ipynb)
    - SHOT in a Python ecosystem (*e.g.*, NumPy, pandas, matplotlib, SciPy, machine learning pipelines, data pipelines)

The notebooks are designed to gradually evolve from basic model construction towards building larger optimization workflows in Python around SHOT. The progression moves from the question

*“How do I solve a model?”*

to the broader and more practical perspective

*“How can I use SHOT together with the Python ecosystem to build larger optimization workflows?”*
